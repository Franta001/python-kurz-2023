```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Načtení dat z CSV souboru
platy = pd.read_csv("platy_2021_02.csv")
platy

```

```python
# Nastavení vhodných hranic skupin histogramu
bins = [0, 30000, 32000, 35000, 37000, 40000, 42000, 45000, 46000, 47000, 49000, 50000, 52000, 55000, 57000, 58000, 60000, 63000, 65000]
```

```python
# Vytvoření histogramu
plt.hist(platy["plat"], bins=bins, edgecolor="black", color = "orange")
plt.title("Rozdělení platů ve firmě (únor 2021)", weight="bold")
plt.xlabel("Plat", color="blue", weight="bold")
plt.xticks(rotation=45)
plt.ylabel("Počet zaměstnanců", color="green", weight="bold")
plt.grid(True)
plt.show()
```
