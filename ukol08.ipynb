```python
import pandas as pd

teplota = pd.read_csv("temperature.csv")
teplota.head()
```

```python
teplota.columns
```

```python
teplota.index
```

```python
praha_data = teplota[teplota["City"] == "Prague"]   # Výpis prvních několika řádků dat z Prahy; zvláštnost spočívá, že se jedná o stupně Fahrenheita a ne Celsia
praha_data.head()
```

```python

vysoka_teplota_data = teplota[teplota["AvgTemperature"] > 80]   # teplota vyšší než 80 stupňů F.
vysoka_teplota_data.head()
```

```python

evropa_vysoka_teplota_data = teplota[(teplota["AvgTemperature"] > 60) & (teplota["Region"] == "Europe")] # Teplota vyšší než 60 stupňů a v regionu Evropa
evropa_vysoka_teplota_data.head()
```

```python

extrem_data = teplota[(teplota["AvgTemperature"] > 80) | (teplota["AvgTemperature"] < -20)]  # teplota vyšší než 80 stupňů nebo menší než -20 stupňů
extrem_data.head()
```

**BONUS**

```python
import pytemperature
import pandas as pd

teplota["AvgTemperatureCelsius"] = pytemperature.f2c(teplota["AvgTemperature"])   # Přidání  sloupce s průměrnou teplotou ve stupních Celsia
teplota.head()
```

```python
vysoka_teplota_celsius_data = teplota[teplota["AvgTemperatureCelsius"] > 30]
vysoka_teplota_celsius_data.head()
```

```python
vys_tep_EU_celsius_data = teplota[(teplota["AvgTemperatureCelsius"] > 15) & (teplota["Region"] == "Europe")]
vys_tep_EU_celsius_data.head()
```

```python
ex_tep_celsius_data = teplota[(teplota["AvgTemperatureCelsius"] > 30) | (teplota["AvgTemperatureCelsius"] < -10)]
ex_tep_celsius_data
```
